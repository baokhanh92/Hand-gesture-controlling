```python
test = pathlib.Path(ROOT, '')
test = list(test.glob('**/*'))
def remove_ipynb(paths_list):
    paths = [pathlib.Path(i) for i in paths_list]
    remove = []
    for path in paths:
        if path.is_dir and path.name == '.ipynb_checkpoints':
            remove.append[path]
    if len(remove) >0:
        shutil.rmtree(path)
        return 'Checkpoint removes'
    else:
        return 'No checkpoint file'
```
